```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/isoforms_results_to_plot.csv')
df.head()
```

```python
df.head(100)
```

```python
#rename grouping1-8 as SC vs committed, NR5A1+ vs rest, TBX19+ vs rest, Cort. vs Mela., POU1F1+ vs rest, Lacto vs Som/Thy, Somato vs Lac/Thy, Thyro vs Som/Lac
rename_dict = {
    'Grouping_1': '1. Stem cells vs committed',
    'Grouping_2': '2. NR5A1+ vs rest',
    'Grouping_3': '3. TBX19+ vs rest',
    'Grouping_4': '4. Corticotrophs. vs Melanotrophs.',
    'Grouping_5': '5. POU1F1+ vs rest',
    'Grouping_6': '6. Lactotrophs vs Som/Thy',
    'Grouping_7': '7. Somatotrophs vs Lac/Thy',
    'Grouping_8': '8. Thyrotrophs vs Som/Lac'
}
df['grouping'] = df['grouping'].map(rename_dict)
```

```python
df["ens_gene"].nunique()
```

```python
df[df["ext_gene"]=="Syngr1"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Categorize by gene_qval threshold
df['category'] = df['gene_qval'] < 0.05

# Count occurrences per grouping
count_data = df.groupby(['grouping', 'category']).size().unstack().fillna(0)
count_data.columns = ['Isoform switching (gene not DE)', 'Isoform switching (gene DE)']

# Plot settings
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
count_data.plot(kind='bar', stacked=False, width=0.8, color=['#8080ff', '#0000ff'], edgecolor='black')

# Customize aesthetics
plt.xlabel("Grouping", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="", fontsize=12)
sns.despine()

#save as png and svg
plt.savefig('isoform_switching_by_grouping.png', dpi=300, bbox_inches='tight')
plt.savefig('isoform_switching_by_grouping.svg', dpi=300, bbox_inches='tight')
```
