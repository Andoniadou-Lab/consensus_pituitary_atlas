```python
#load /Users/k23030440/Library/CloudStorage/OneDrive-King\'sCollegeLondon/PhD/Year_two/Aim\ 1/epitome/data/aging/v_0.01/aging_genes.csv
aging = pandas.read_csv('/Users/k23030440/epitome_code/epitome/data/aging/v_0.02/aging_genes.csv')
#keep where cell_type is Stem_Cells, where adj.P.Val is <0.05 and logFC is >0
aging = aging[(aging['cell_type'] == 'Stem_cells') & (aging['adj.P.Val'] < 0.05) & (aging['logFC'] > 0)]
#fisher test with genes
from scipy.stats import fisher_exact
#make a 2x2 table
#in sen mayo and in aging
in_senmayo_and_aging = len(set(genes) & set(aging['genes']))
#in sen mayo and not in aging
in_senmayo_and_not_aging = len(genes) - in_senmayo_and_aging
#not in sen mayo and in aging
not_in_senmayo_and_in_aging = len(aging) - in_senmayo_and_aging
#not in sen mayo and not in aging
not_in_senmayo_and_not_aging = 21075 - (in_senmayo_and_aging + in_senmayo_and_not_aging + not_in_senmayo_and_in_aging)
table = [[in_senmayo_and_aging, in_senmayo_and_not_aging], [not_in_senmayo_and_in_aging, not_in_senmayo_and_not_aging]]
oddsratio, pvalue = fisher_exact(table, alternative='greater')
oddsratio, pvalue
```

```python
print(f"Number of genes in both SenMayo and aging: {in_senmayo_and_aging}")
print(f"These are the genes: {set(genes) & set(aging['genes'])}")
```
