```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/pituitary_atlas/Pituitary markers from publications.xlsx')
#if Gene column empty remove row
df = df.dropna(subset=['Gene'])
#for those rows where Publication and Cluster_broad columns are the same, concatenate Gene entries
df = df.groupby(["Publication", "Cluster_broad"]).agg({"Gene": ", ".join}).reset_index()
#from each Gene column, extract comma separated list of values and keep only unique entries
df['Gene'] = df['Gene'].apply(lambda x: list(set(x.split(", "))))
#Now finally, group.by Cluster_broad and once again join all Gene lists by appending. Then for each unique Cluster_broad print out the entries from those lists that occur at least 3 times
for cluster in df['Cluster_broad'].unique():
  #get all rows where Cluster_broad is equal to cluster
  cluster_df = df[df['Cluster_broad'] == cluster]
  all_genes = []
  for genes in cluster_df['Gene']:
    all_genes.extend(genes)
  occurs_more = {}
  for gene in all_genes:
    if gene in occurs_more:
      occurs_more[gene] += 1
    else:
      occurs_more[gene] = 1
  print("*****************************")
  print(cluster)
  for gene, count in occurs_more.items():
    if count >= 2:
      print(gene)
  print("*****************************")
```
