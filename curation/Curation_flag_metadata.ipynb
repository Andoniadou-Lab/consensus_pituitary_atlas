```python
#turn these to integers based on the first number character 'Study_Seurat_version', 'Study_CR_version' using regex
df['Study_Seurat_version_int'] = df['Study_Seurat_version'].str.extract('(\d+)').astype('Int64')
df['Study_CR_version_int'] = df['Study_CR_version'].str.extract('(\d+)').astype('Int64')
```

```python
#dropna
df_versions = df.dropna(subset=['Study_CR_version_int'])

df_versions ['Study_CR_version_int']

df_versions ['Author_year'] =df_versions ['Author'].str.extract('(\d{4})')
df_versions ['Author_year']


#print where year is > 2023 and ['Study_CR_version_int'] < 5
print(df_versions [(df_versions ['Author_year'].astype('Int64') > 2023) & (df_versions ['Study_CR_version_int'] < 7) & (df_versions ['Study_CR_version_int'] > 2)]['Author'].unique())

#print where year is > 2023 and ['Study_CR_version_int'] < 5
print(df_versions [(df_versions ['Author_year'].astype('Int64') > 2023) & (df_versions ['Study_CR_version_int'] < 7) & (df_versions ['Study_CR_version_int'] > 2)]['Author'].nunique())
```
