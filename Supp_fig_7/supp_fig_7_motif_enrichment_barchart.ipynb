```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import numpy as np

#load /content/multimodal_hits_combined.csv
import pandas as pd
df = pd.read_csv('/content/multimodal_hits.csv')
#keep where grouping is grouping_1_up
df = df[df['grouping'] == 'grouping_1_up']
#order by atac_fold_enrichment
df = df.sort_values(by=['atac_fold_enrichment'], ascending=False)

atac_fold_enrichment = df['atac_fold_enrichment']
n_tfs = len(atac_fold_enrichment)

# Set a sans-serif font for a publication-ready look
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica', 'Liberation Sans', 'DejaVu Sans', 'Bitstream Vera Sans']
plt.rcParams['font.size'] = 10 # Adjust base font size if needed
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['axes.titlesize'] = 14

plt.figure(figsize=(7.5, 4))

#upper the gene names
df["gene"] = df["gene"].str.upper()

sns.barplot(
    data=df,
    x="gene",
    y='atac_fold_enrichment',
    color='#0000ff', # Use your generated list of colors
    edgecolor='black', # Add a subtle edge for better definition
    linewidth=0.7 # Line width of the edge
)

# Rotate x-axis ticks for readability with many gene names

plt.xticks(rotation=90)
#larger ticks

# Rename y-label for clarity and remove x-label as it's often redundant with rotated ticks
plt.xlabel('')
plt.ylabel('Motif Fold-Enrichment')


# Set y-axis limits
plt.ylim(1, 2)

# Add a title if appropriate, or keep it concise for figure legends
# plt.title('Motif Fold-Enrichment Across Genes')

# Ensure layout is tight to prevent labels from being cut off
plt.tight_layout()

# Save figures in high resolution
plt.savefig('atac_fold_enrichment.png', dpi=600, bbox_inches='tight')
plt.savefig('atac_fold_enrichment.svg', format='svg', dpi=600, bbox_inches='tight')

plt.show() # Display the plot (optional, for direct viewing)
plt.show()
```
